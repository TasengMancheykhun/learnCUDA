```python
%%writefile prog.cu
#include<stdio.h>
#define N 100

__global__ void countevenodd(int *d_input, int *d_even, int *d_odd)
{
  int idx = blockIdx.x * blockDim.x + threadIdx.x;

  if (idx<N)
  {
      int evenflag=0, oddflag=0;

      if (d_input[idx]%2==0)
      {
          evenflag=1;
      }
      else
      {
          oddflag=1;
      }

      d_even[idx]=evenflag;
      d_odd[idx]=oddflag;
  }

}


int main()
{

  int h_array[N];
  int *d_array;
  int *d_evencount, *d_oddcount;
  int h_evencount[N], h_oddcount[N];
  int c_even=0, c_odd=0;

  for (int i=0;i<100;i++)
  {
    h_array[i]=i;
  }

  cudaMalloc(&d_array, N*sizeof(int));
  cudaMalloc(&d_evencount, N*sizeof(int));
  cudaMalloc(&d_oddcount, N*sizeof(int));


  cudaMemcpy(d_array, h_array, N*sizeof(int), cudaMemcpyHostToDevice);

  countevenodd<<<1,N>>>(d_array, d_evencount, d_oddcount);

  cudaMemcpy(h_evencount, d_evencount, N*sizeof(int), cudaMemcpyDeviceToHost);
  cudaMemcpy(h_oddcount, d_oddcount, N*sizeof(int), cudaMemcpyDeviceToHost);


  for(int i=0;i<N;i++)
  {
       c_even+=h_evencount[i];
       c_odd+=h_oddcount[i];
  }



  printf("Count of EVEN numbers between 0 and 99 is: %d \n", c_even);

  printf("Count of ODD numbers between 0 and 99 is: %d \n", c_odd);

  cudaFree(d_array);
  cudaFree(d_evencount);
  cudaFree(d_oddcount);

  return 0;
}
```
