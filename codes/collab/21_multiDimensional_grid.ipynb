```python
%%writefile prog.cu

#include<stdio.h>

__global__ void kernel(){
  int bx =blockIdx.x;
  int by =blockIdx.y;
  int tx =threadIdx.x;
  int ty =threadIdx.y;
  int gd=gridDim.x;

  // int gd=a.x;               // cannot do it
  int bd=blockDim.x;

  // Calculate the global index consdering both grid and block dimensions

  int idx =(by * gridDim.x+bx)*(blockDim.x*blockDim.y)+(ty*blockDim.x)+tx;

  printf("Block (%d, %d), Thread (%d ,%d), Griddim.x (%d), Blockdim.x(%d), Global Index : %d\n", bx, by, tx, ty, gd, bd, idx);

}


int main()
{

dim3 gridDim(5,1);
dim3 blockDim(5,1);

//dim3 a(5,1);                // this is allowed
//dim3 b(5,1);

kernel<<<gridDim,blockDim>>>();
// kernel<<<a,b>>>();            // this is allowed

cudaDeviceSynchronize();


  return 0;
}
```
