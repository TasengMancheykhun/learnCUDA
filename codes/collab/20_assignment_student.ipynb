```python
%%writefile prog.cu

#include<stdio.h>


__device__ int findtotal(int marks[3])
   {
        int sum=0;
        for (int i=0;i<3;i++)
        {
            sum += marks[i];
        }

        return sum;
   }


__global__ void result(int *d_marks, int *d_result)
{
    int idx = threadIdx.x + blockIdx.x * blockDim.x;

    if (idx<5)
    {
      int marks[3];
      for (int i=0; i<3; i++)
      {

        marks[i] = d_marks[idx*3+i];

      }

     d_result[idx] = findtotal(marks);

    }

}


int main()
{

  int h_marks[5][3] = {
          {60, 70, 80},
          {45, 55, 65},
          {80, 90, 95},
          {35, 40, 50},
          {70, 80, 85}
  };


  int *d_marks, *d_result;
  int h_output[5];

  cudaMalloc(&d_marks, 5*3*sizeof(int));
  cudaMalloc(&d_result, 5*sizeof(int));

  cudaMemcpy(d_marks, h_marks, 5*3*sizeof(int), cudaMemcpyHostToDevice);

  result<<<1,5>>>(d_marks, d_result);

  cudaMemcpy(h_output, d_result, 5*sizeof(int), cudaMemcpyDeviceToHost);

  printf("Total of each student is: \n");

  for (int i=0; i<5; i++)
  {
    printf("%d ",h_output[i]);
  }
  printf("\n");

  cudaFree(d_marks);
  cudaFree(d_result);

  return 0;
}
```
