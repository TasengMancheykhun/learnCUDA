```python
%%writefile prog.cu
#include <stdio.h>

__device__ int square(int x)
{
  return x*x;
}

__global__ void squareKernel(int *d_input, int *d_output)
{
  int idx = threadIdx.x + blockIdx.x * blockDim.x;

  if (idx<5)
  {
    d_output[idx] = square(d_input[idx]);
  }
}


int main()
{
  int h_input[] = {1,2,3,4,5};
  int h_output[5];

  int *d_input, *d_output;

  cudaMalloc(&d_input, 5*sizeof(int));
  cudaMalloc(&d_output, 5*sizeof(int));

  cudaMemcpy(d_input, h_input, 5*sizeof(int), cudaMemcpyHostToDevice);

  squareKernel<<<1,5>>>(d_input, d_output);

  cudaMemcpy(h_output,d_output, 5*sizeof(int), cudaMemcpyDeviceToHost);

  printf("Squares of the input array: \n");
  for (int i=0; i<5; i++)
  {
    printf("%d ",h_output[i]);
  }
  printf("\n");

  cudaFree(d_input);
  cudaFree(d_output);

}
```
