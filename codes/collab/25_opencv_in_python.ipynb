```python
import cv2
from google.colab.patches import cv2_imshow

image = cv2.imread('parrot.jpg')

print("Size: ", image.shape[:2])  #Height and width
print("Mode:", "BGR")  #OpenCV loads images in BGR format by default
print("Format: JPEG")   #format information usually comes from file type (if known)

#Resize the image to a smaller size
width = 500  #Set the desired width
height = int(image.shape[0]*(width/image.shape[1]))   #Maintain the aspect ratio
resized_image = cv2.resize(image, (width, height))

# #Show the resized image in a window
# cv2.imshow('parrot', resized_image)
# cv2.waitKey(0) #Wait for a key press to close the window
# cv2.destroyAllWindows()

#convert the image to grayscale
gray_image = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)

#save the new grayscale image
cv2.imwrite('parrot-gray.jpg', gray_image)

# #display the grayscale image in a window
# cv2_imshow('grayscale parrot', gray_image)
# cv2.waitKey(0)         #Wait for a key to close the window
# cv2.destroyAllWindows()
```
