```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

from PIL import Image

uploaded = files.upload()
pic = Image.open('parrot.jpg')

print(type(pic))

#View image
plt.imshow(pic)
plt.axis('off')  #Turn off axis numbers and ticks
plt.show()


#convert image to a numpy array
pic_arr = np.asarray(pic)
print(type(pic_arr))

print(pic_arr.shape)

plt.imshow(pic_arr)
plt.show()
```

```python
#Remove red pixels

pic_gb = pic_arr.copy()
pic_gb[:,:,0] = 0 #set the red channel to 0
print(pic_gb.shape)

plt.imshow(pic_gb)
plt.show()
```

```python
pic_gb = pic_arr.copy()
pic_gb[:,:,1] = 0 #set the green channel to 0
print(pic_gb.shape)

plt.imshow(pic_gb)
plt.show()
```

```python
pic_gb = pic_arr.copy()
pic_gb[:,:,2] = 0    #set the blue channel to 0
print(pic_gb.shape)

plt.imshow(pic_gb)
plt.show()
```

```python
# custom multiple colors
pic_custom = pic_arr.copy()
pic_custom[:,:,1]=0      # remove green
pic_custom[:,:,0] = np.clip(pic_custom[:,:,0] + 50, 0, 255)  #boost red

pic_custom[:,:,2] = np.clip(pic_custom[:,:,2] + 30, 0, 255)   #boost blue

plt.imshow(pic_custom)
plt.show()
```

```python
pic_custom = pic_arr[100:200, 150:250, 1]
plt.imshow(pic_custom)
plt.axis('off')
plt.show()
```

```python
#using the pillow package: first do pip install pillow
from PIL import Image

image = Image.open('parrot.jpg')
print(image.size, image.mode, image.format)

image.show()

#convert the image to grayscale
gray_image = image.convert("L")

#Save the new grayscale image
gray_image.save('parrot-gray.jpg')

plt.imshow(gray_image)
plt.axis('off')
plt.show()
```
