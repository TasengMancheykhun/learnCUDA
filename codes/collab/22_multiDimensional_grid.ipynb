```python
%%writefile prog.cu

#include<stdio.h>

__global__ void kernel(){
  int bx =blockIdx.x;
  int by =blockIdx.y;
  int tx =threadIdx.x;
  int ty =threadIdx.y;

  // Calculate the global index consdering both grid and block dimensions

  int idx =(by * gridDim.x+ bx )*(blockDim.x*blockDim.y)+(ty*blockDim.x)+tx;

  printf("Block (%d, %d), Thread (%d ,%d), Griddim.x(%d), Blockdim.x(%d), Griddim.y(%d), Blockdim.y(%d), Global Index : %d\n", bx, by, tx, ty, gridDim.x, blockDim.x, gridDim.y, blockDim.y, idx);

}


int main()
{

dim3 gridDim(2,2);
dim3 blockDim(2,2);


kernel<<<gridDim,blockDim>>>();

cudaDeviceSynchronize();


  return 0;
}
```
